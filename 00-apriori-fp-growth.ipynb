"룰 기반 알고리즘"
-단점: 메모리를 많이 차지, 생성된 규칙이 많아 막상 적용하기가 어려움

```python
import mlxtend
import numpy as np
import pandas as pd
```

```python
# 데이터셋을 생성
data = np.array([
    ['우유', '기저귀', '쥬스'], # 0번 거래
    ['양상추', '기저귀', '맥주'], # 1번 거래
    ['우유', '양상추', '기저귀', '맥주'], # 2번 거래
    ['양상추', '맥주'] # 3번 거래
])
```

## Apriori 알고리즘

```python
from mlxtend.preprocessing import TransactionEncoder # TransactionEncoder 불러오기

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# 연습
pd.DataFrame(te_ary)
```

```python
# 연습
te.columns_
```

```python
%%time
from mlxtend.frequent_patterns import apriori # apriori 라이브러리 불러오기

apriori(df, min_support=0.5, use_colnames=True) # 최소 지지도 설정
```

## FP-Growth 알고리즘

```python
# 전처리 단계는 동일
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
%%time
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.5, use_colnames=True) # fpgrowth -> apriori보다 속도가 빠름 (장점)
```

```python
# 추가
from mlxtend.frequent_patterns import association_rules

association = fpgrowth(df, min_support=0.1, use_colnames=True)
association_rules(association, metric = 'confidence', min_threshold=0.5, support_only=True)
```
